```python
import matplotlib.pyplot as plt

# Data for plotting
cores = [1, 2, 3, 4, 5, 6]

# Times for pthread
times_pthread = [101.081797/101.081797, 101.081797/47.039247, 101.081797/36.078549, 101.081797/27.901165, 101.081797/22.196732, 101.081797/19.767640]


# Times for hybrid
times_hybrid = [15.935971/15.935971, 15.935971/9.764404, 15.935971/7.719552, 15.935971/6.703417, 15.935971/6.192937, 15.935971/5.715852]


# Optimal scaling (assuming linear scaling with 1 core taking the time of the first pthread time)
times_optimal = [1,2,3,4,5,6]


# Creating the line plot
plt.figure(figsize=(10, 6))
plt.plot(cores, times_pthread, marker='o', label='pthread')
plt.plot(cores, times_hybrid, marker='s', label='hybrid')
plt.plot(cores, times_optimal, marker='^', linestyle='--', label='optimal (linear scaling)')

plt.xlabel('Number of Cores')
plt.ylabel('Execution Time (seconds)')
plt.title('Scalability Comparison')
plt.xticks(cores)
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

```
